```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Perceptron 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
or_data = pd.DataFrame()     
and_data = pd.DataFrame()     
xor_data = pd.DataFrame()
    
or_data['input_1'] = [1,1,0,0]     
or_data['input_2'] = [1,0,1,0]     
or_data['output'] = [1,1,1,0]     

and_data['input_1'] = [1,1,0,0]     
and_data['input_2'] = [1,0,1,0]     
and_data['output'] = [1,0,0,0]     

xor_data['input_1'] = [1,1,0,0]     
xor_data['input_2'] = [1,0,1,0]     
xor_data['output'] = [0,1,1,0]     
```

```python
or_data
```

```python
and_data
```

```python
xor_data
```

```python
sns.scatterplot(x=and_data['input_1'],y=and_data['input_2'],hue=and_data['output'],s=200)
```

```python
sns.scatterplot(x=or_data['input_1'],y=or_data['input_2'],hue=or_data['output'],s=200)
```

```python
sns.scatterplot(x=xor_data['input_1'],y=xor_data['input_2'],hue=xor_data['output'],s=200)
```

```python
model1 = Perceptron()
model2 = Perceptron()
model3 = Perceptron()
```

```python
model1.fit(and_data.iloc[:,0:2].values, and_data.iloc[:,-1].values)
model2.fit(or_data.iloc[:,0:2].values, or_data.iloc[:,-1].values)
model3.fit(xor_data.iloc[:,0:2].values, xor_data.iloc[:,-1].values)
```

```python
x = np.linspace(-1,1,5)
y = -x + 1
```

```python
plt.plot(x,y)
sns.scatterplot(x=and_data['input_1'],y=and_data['input_2'],hue=and_data['output'],s=200)
```

```python
x = np.linspace(-1,1,5)
y = -x + 0.5
```

```python
plt.plot(x,y)
sns.scatterplot(x=or_data['input_1'],y=or_data['input_2'],hue=or_data['output'],s=200)
```

```python
from mlxtend.plotting import plot_decision_regions 
```

```python
plot_decision_regions(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values,clf=model3,legend=2)
```

- **No clear boundary here**
- **Thus, Perceptron fails on Non-Linear data**
